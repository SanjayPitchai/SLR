Training RandomForestClassifier() model

```python
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
data_dict = pickle.load(open('./data1.pickle', 'rb'))
```

```python
# Pad sequences to a fixed length
max_length = 100
padded_data = pad_sequences(data_dict['data'], maxlen=max_length, padding='post', dtype='float32', truncating='post')
```

```python
data = np.asarray(padded_data)
labels = np.asarray(data_dict['labels'])
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels,random_state=25)
```

```python
model = RandomForestClassifier()
# epochs = 20
# for epoch in range(epochs):
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
score = accuracy_score(y_predict, y_test)

print('{}% of samples were classified correctly !'.format(score * 100))
```

```python
acc = accuracy_score(y_predict, y_test)
prec = precision_score(y_predict, y_test, average='macro',zero_division=1)
rec = recall_score(y_predict, y_test, average='macro',zero_division=1)
f1 = f1_score(y_predict, y_test, average='macro')
print("Accuracy:", acc)
print("Precision:", prec)
print("Recall:", rec)
print("F1-score:", f1)
```

```python
f = open('model1.p', 'wb')
pickle.dump({'model': model}, f)
f.close()
```

Training SVC model for comparision

```python
import pickle
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the data from the pickle file
data_dict = pickle.load(open('./data.pickle', 'rb'))

# Pad sequences to a fixed length
max_length = 100  
padded_data = pad_sequences(data_dict['data'], maxlen=max_length, padding='post', dtype='float32', truncating='post')

# Convert data and labels to NumPy arrays
data = np.asarray(padded_data)
labels = np.asarray(data_dict['labels'])

x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels, random_state=25)

model = SVC()
# training data
model.fit(x_train, y_train)
# test data
y_predict = model.predict(x_test)


accuracy = accuracy_score(y_test, y_predict)
precision = precision_score(y_test, y_predict, average='macro', zero_division=1)
recall = recall_score(y_test, y_predict, average='macro', zero_division=1)
f1 = f1_score(y_test, y_predict, average='macro')
print('Accuracy: {:.4f}'.format(accuracy))
print('Precision: {:.4f}'.format(precision))
print('Recall: {:.4f}'.format(recall))
print('F1-score: {:.4f}'.format(f1))

# Save the model
with open('model_svc.p', 'wb') as f:
    pickle.dump({'model': model}, f)
```
